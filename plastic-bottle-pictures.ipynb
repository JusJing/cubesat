```python
from imageai.Detection import ObjectDetection
import os

execution_path = "Plastic-Bottles-Dataset/"
output_path = "Plastic-Bottles-Dataset-output/"
if not os.path.exists(output_path):
    os.makedirs(output_path)

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(os.getcwd(),"yolov3.pt"))
detector.loadModel()

dir_list = os.listdir(execution_path)
for d in dir_list:
    abs_dir = os.path.join(execution_path, d + "/")
    if os.path.isdir(abs_dir):
        img_list = os.listdir(abs_dir)
        for file in img_list:
            if file.endswith(".jpg"):
                detections, objects_path = detector.detectObjectsFromImage(input_image=os.path.join(abs_dir, file), output_image_path=os.path.join(output_path, file), minimum_percentage_probability=30, extract_detected_objects=True)

                for eachObject, eachObjectPath in zip(detections, objects_path):
                    print(eachObject["name"] , " : " , eachObject["percentage_probability"], " : ", eachObject["box_points"] )
                    print("Object's image saved in " + eachObjectPath)
                    print("--------------------------------")
```
